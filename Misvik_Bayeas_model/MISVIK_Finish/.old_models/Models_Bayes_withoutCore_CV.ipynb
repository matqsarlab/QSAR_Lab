```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import plotly.figure_factory as ff
from sklearn.metrics import classification_report
import itertools
from QSAR_Lab.Misvik_Bayeas_model.models import CoreModel, CombinationModel
from QSAR_Lab.Misvik_Bayeas_model.comparison import observe, compare
from QSAR_Lab.Misvik_Bayeas_model.pre_crossValidation import Selector
from IPython.display import display


from sklearn.naive_bayes import GaussianNB
```

```python
my_model = GaussianNB()
```

# OX A549

```python
train_a549 = pd.read_excel("OX/Data/a549_data.xlsx", sheet_name=1, index_col=0)
test_a549 = pd.read_excel("OX/Data/a549_data.xlsx", sheet_name=2, index_col=0)
toxic_a549 = pd.read_excel("OX/Data/a549_data.xlsx", sheet_name=3, index_col=0)
```

```python
# Ilosc toksycznych NMs w zbiorze testowym
tox_in_test = 2
```

```python
external_data = Selector(
    my_model, train_a549, test_a549, toxic_a549, tox_in_test
).create_data_set()
```

```python
lista_tabel = []
for key in external_data:
    train, test, toxic, external = external_data[key]
    a549 = CombinationModel(my_model, train, test, toxic, tox_in_test)
    a549.create_models
    table = compare(pd.DataFrame(external).T,a549.predict(pd.DataFrame(external.drop("y")).T))
    lista_tabel.append(table)
```

```python
for i in lista_tabel:
    display(i)
```
